```python
merged = pd.merge(trades, sentiment, on='date', how='left')
```

```python
merged[['date', 'Closed_PnL', 'classification']].head()
```

```python
merged['win'] = merged['Closed_PnL'] > 0
```

```python
daily_pnl = merged.groupby(['Account', 'date'])['Closed_PnL'].sum().reset_index()
```

```python
win_rate = merged.groupby('Account')['win'].mean().reset_index()
```

```python
trades_per_day = merged.groupby('date').size().reset_index(name='trade_count')
```

```python
avg_trade_size = merged.groupby('Account')['Size_USD'].mean().reset_index()
```

```python
merged['Side'].value_counts()
```

```python
long_short_ratio = merged.groupby('classification')['Side'].value_counts(normalize=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='classification', y='Size_USD', data=merged)
plt.xticks(rotation=45)
plt.title("Average Trade Size vs Market Sentiment")
plt.show()
```
